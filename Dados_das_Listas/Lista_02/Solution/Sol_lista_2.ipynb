# **Validação Cruzada**

```python
# Validacao Cruzada
import random
def kfoldcv(indices, k = 10, seed = 42):
    
    size = len(indices)
    subset_size = round(size / k)
    random.Random(seed).shuffle(indices)
    subsets = [indices[x:x+subset_size] for x in range(0, len(indices), subset_size)]
    kfolds = []
    for i in range(k):
        test = subsets[i]
        train = []
        for subset in subsets:
            if subset != test:
                train.append(subset)
        kfolds.append((train,test))
        
    return kfolds
```

```python
# Concatena os Indices Gerados em kfoldcv
# Nível de índices: iteracao -> x_train (0) ou test (1)
def concat_fold(array_fold):
  K = array_fold 
  folds = []
  for i in range(len(K)):
    train = K[i][0]
    test  = K[i][1]
    x_train = []
    for j in train: x_train += j
    folds.append([x_train,test])
  return folds 
```

```python
# Retorna um df de iteracoes dos kfolds
def df_folds(dataframe, kfold):

  k_in = concat_fold(kfold)
  df_train = []
  df_test = []
  for i in range(len(k_in)):
      x_index = k_in[i][0]
      df_train.append(dataframe.loc[x_index])

      t_index = k_in[i][1]
      df_test.append(dataframe.loc[t_index])

  return df_train, df_test
```

#**Normalizando os Dados**

```python
# Função que normaliza um único dataframe
def data_norm(dataframe,col_y=None):
  if col_y != None:
    dt_y = dataframe[col_y].values
    dt = dataframe.drop(col_y,axis=1)
    norm = (dt-dt.mean())/dt.std()
    norm[col_y] = dt_y
  else:
    dt = dataframe
    norm = (dt-dt.mean())/dt.std()
  return norm

# Função que 'desnormaliza' um único dataframe
def data_norm_inv(dataframe_norm,dataframe,col_y=None):
  if col_y != None:
    df = dataframe.drop(col_y,axis=1) 
    df_norm_y = dataframe_norm[col_y].values
    df_norm = dataframe_norm.drop(col_y,axis=1)
    inv = df_norm*df.std() + df.mean()
    inv[col_y] = df_norm_y

  else:
    df = dataframe
    df_norm = dataframe_norm
    inv = df_norm*df.std() + df.mean()
  return inv
```

```python
# Função Para Normalizar n dataframes
def full_data(full_dataframe, col_y=None):
  df = full_dataframe
  vec_norm = []
  for i in range(len(df[0])):
    vec_norm.append( data_norm(df[0][i],col_y) )
  return vec_norm

# Função Para 'Desormalizar' n dataframes
def full_data_inv(full_dataframe_norm, full_dataframe, col_y=None):
  df_norm = full_dataframe_norm
  df = full_dataframe
  vec_x = []
  for i in range(len(df[0])):
    vec_x.append( data_norm_inv(df_norm[i], df[0][i], col_y) )
  return vec_x
```

# **Regressão Logística via SGD**

```python
# Função sigmóide
def sigmoid(z):
    return 1 / (1 + np.exp(-z-1e-9))

# Função de custo -- 0bs: foi adicionado '1e-9' no log para evitar overflow
def cost_function(X, y, theta):
    h = sigmoid(X @ theta)
    cost = (y.T @ np.log(h)) + ((1 - y).T @ np.log(1 - h + 1e-9))
    return -cost

# Função de gradiente descendente estocástico
def sgd_logi(X, y, alpha, num_iter):
    m, n = X.shape
    theta = np.zeros((n, 1))
    for i in range(num_iter):
        rand_idx = np.random.randint(m)
        xi = X[rand_idx, :].reshape(1, -1)
        yi = y[rand_idx].reshape(1, -1)
        hi = sigmoid(xi @ theta)
        ei = hi - yi
        grad = xi.T @ (ei)
        theta -= alpha * grad
    return theta,ei
```

# **Análise do Discriminante Gaussiano**

```python
def AD_gauss(train_df,test_df): 
  # Dividir os dados de treino em duas classes com base na coluna de saída binária
  classe1 = train_df[train_df.iloc[:, -1] == 0]
  classe2 = train_df[train_df.iloc[:, -1] == 1]

  # Calcular a média e a matriz de covariância de cada classe
  mean1 = np.mean(classe1.iloc[:, :-1], axis=0)
  cov1 = np.cov(classe1.iloc[:, :-1].T)
  mean2 = np.mean(classe2.iloc[:, :-1], axis=0)
  cov2 = np.cov(classe2.iloc[:, :-1].T)

  # Calcular a matriz de covariância comum
  cov = ((len(classe1) - 1) * cov1 + (len(classe2) - 1) * cov2) / (len(train_df) - 2)

  # Calcular as matrizes inversas e determinantes das matrizes de covariância
  inv_cov = np.linalg.inv(cov)
  det_cov = np.linalg.det(cov)
  inv_cov1 = np.linalg.inv(cov1)
  det_cov1 = np.linalg.det(cov1)
  inv_cov2 = np.linalg.inv(cov2)
  det_cov2 = np.linalg.det(cov2)

  # Calcular as probabilidades de cada classe
  prob1 = len(classe1) / len(train_df)
  prob2 = len(classe2) / len(train_df)

  # Definir a função discriminante para cada classe
  def discriminante(x):
      g1 = -0.5 * (x - mean1).dot(inv_cov1).dot((x - mean1).T) - 0.5 * np.log(det_cov1) + np.log(prob1)
      g2 = -0.5 * (x - mean2).dot(inv_cov2).dot((x - mean2).T) - 0.5 * np.log(det_cov2) + np.log(prob2)
      return 1 if g2 > g1 else 0

  # Aplicar a função discriminante aos dados de treino e teste
  y_train_pred = np.apply_along_axis(discriminante, 1, train_df.iloc[:, :-1])
  y_test_pred = np.apply_along_axis(discriminante, 1, test_df.iloc[:, :-1])

  # Comparar as previsões com a saída real e calcular a acurácia para treino e teste
  train_accuracy = np.sum(y_train_pred == train_df.iloc[:, -1]) / len(train_df)
  test_accuracy = np.sum(y_test_pred == test_df.iloc[:, -1]) / len(test_df)

  #print('Acurácia de Treino:', train_accuracy)
  #print('Acurácia de Teste:', test_accuracy)
  return y_train_pred, y_test_pred, train_accuracy, test_accuracy
```

# **Naive Bayes**

```python
def NB(train,test):
  # Separando as classes
  classe0 = train[train.iloc[:, -1] == 0]
  classe1 = train[train.iloc[:, -1] == 1]

  # Calculando as médias e desvios padrões de cada classe
  media0 = np.mean(classe0.iloc[:, :-1], axis=0)
  desvio0 = np.std(classe0.iloc[:, :-1], axis=0)
  media1 = np.mean(classe1.iloc[:, :-1], axis=0)
  desvio1 = np.std(classe1.iloc[:, :-1], axis=0)

  # Função de densidade de probabilidade gaussiana
  def pdf(x, media, desvio):
      return 1/(desvio*np.sqrt(2*np.pi))*np.exp(-((x-media)**2)/(2*desvio**2))

  # Função de classificação Naive Bayes gaussiano
  def classifica(x):
      prob0 = np.prod(pdf(x[:-1], media0, desvio0))*len(classe0)/len(train)
      prob1 = np.prod(pdf(x[:-1], media1, desvio1))*len(classe1)/len(train)
      if prob0 > prob1:
          return 0
      else:
          return 1

  # Classificando os dados de teste
  y_pred = []
  for i in range(len(test)):
      y_pred.append(classifica(test.iloc[i]))
  y_pred = np.array(y_pred)

  # Calculando a acurácia da classificação
  accuracy = np.sum(y_pred == test.iloc[:, -1]) / len(test)
  #print("Acurácia: {:.2f}%".format(accuracy*100))
  return y_pred, accuracy
```

**Cria uma Função para Logist. Multiclasse**

```python
def log_mult(train,test):
  # Converter as colunas em arrays numpy
  X_train = train.iloc[:, :-1].to_numpy()
  y_train = train.iloc[:, -1].to_numpy()
  X_test = test.iloc[:, :-1].to_numpy()
  y_test = test.iloc[:, -1].to_numpy()
  # Função de ativação sigmoid
  def sigmoid(z):
      return 1 / (1 + np.exp(-z))

  # Função de custo
  def cost_function(X, y, theta):
      m = len(y)
      h = sigmoid(np.dot(X, theta))
      J = (-1/m) * (np.dot(y.T, np.log(h)) + np.dot((1 - y).T, np.log(1 - h)))
      return J

  # Gradiente descendente
  def gradient_descent(X, y, theta, alpha, num_iters):
      m = len(y)
      J_history = np.zeros(num_iters)
      for i in range(num_iters):
          h = sigmoid(np.dot(X, theta))
          theta = theta - (alpha/m) * np.dot(X.T, (h - y))
          J_history[i] = cost_function(X, y, theta)
      return theta, J_history

  # Adicionar coluna de 1s para o termo de bias
  X_train = np.c_[np.ones((len(X_train), 1)), X_train]
  X_test = np.c_[np.ones((len(X_test), 1)), X_test]

  # Número de classes
  num_classes = len(np.unique(y_train))

  # Treinamento
  all_theta = np.zeros((num_classes, X_train.shape[1]))
  for c in range(num_classes):
      y_train_c = np.where(y_train == c, 1, 0)
      theta = np.zeros(X_train.shape[1])
      theta, J_history = gradient_descent(X_train, y_train_c, theta, alpha=0.1, num_iters=1000)
      all_theta[c,:] = theta

  # Previsão
  h = sigmoid(np.dot(X_test, all_theta.T))
  y_pred = np.argmax(h, axis=1)

  # Acurácia
  accuracy = np.mean(y_pred == y_test)
  #print("Acurácia:", accuracy)
  return y_pred,accuracy,J_history
```
